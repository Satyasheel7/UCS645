```python
%%bash
cat > deviceQuery.cu << 'EOF'
#include <stdio.h>
#include <cuda_runtime.h>

int main() {
    int deviceCount = 0;
    cudaGetDeviceCount(&deviceCount);

    if (deviceCount == 0) {
        printf("No CUDA devices found!\n");
        return 1;
    }

    printf("Number of CUDA devices: %d\n\n", deviceCount);

    int device = 0;
    cudaSetDevice(device);

    cudaDeviceProp prop;
    cudaGetDeviceProperties(&prop, device);

    printf("========== CUDA Device Query ==========\n\n");

    printf("Device %d: %s\n", device, prop.name);
    printf("========================================\n\n");

    printf("1. ARCHITECTURE AND COMPUTE CAPABILITY:\n");
    printf("   Compute Capability: %d.%d\n\n", prop.major, prop.minor);

    printf("2. MAXIMUM BLOCK DIMENSIONS:\n");
    printf("   Max Threads per Block (X dimension): %d\n", prop.maxThreadsDim[0]);
    printf("   Max Threads per Block (Y dimension): %d\n", prop.maxThreadsDim[1]);
    printf("   Max Threads per Block (Z dimension): %d\n", prop.maxThreadsDim[2]);
    printf("   Max Threads per Block (total): %d\n\n", prop.maxThreadsPerBlock);

    printf("3. MAXIMUM GRID DIMENSIONS:\n");
    printf("   Max Grid Size (X dimension): %d\n", prop.maxGridSize[0]);
    printf("   Max Grid Size (Y dimension): %d\n", prop.maxGridSize[1]);
    printf("   Max Grid Size (Z dimension): %d\n\n", prop.maxGridSize[2]);

    printf("4. GLOBAL MEMORY:\n");
    printf("   Total Global Memory: %zu MB (%zu bytes)\n",
           prop.totalGlobalMem / (1024 * 1024), prop.totalGlobalMem);
    printf("   Global Memory per Block: %zu bytes\n\n", prop.totalGlobalMem);

    printf("5. SHARED MEMORY:\n");
    printf("   Shared Memory per Block: %zu KB (%zu bytes)\n",
           prop.sharedMemPerBlock / 1024, prop.sharedMemPerBlock);
    printf("   Shared Memory per Multiprocessor: %zu KB\n\n",
           prop.sharedMemPerMultiprocessor / 1024);

    printf("6. CONSTANT MEMORY:\n");
    printf("   Total Constant Memory: %zu KB (%zu bytes)\n\n",
           prop.totalConstMem / 1024, prop.totalConstMem);

    printf("7. WARP SIZE:\n");
    printf("   Warp Size: %d threads\n\n", prop.warpSize);

    printf("8. MULTIPROCESSORS:\n");
    printf("   Number of Multiprocessors: %d\n", prop.multiProcessorCount);
    printf("   Max Threads per Multiprocessor: %d\n\n", prop.maxThreadsPerMultiProcessor);

    printf("9. MEMORY SPECIFICATIONS:\n");
    printf("   Memory Clock Rate: %d MHz\n", prop.memoryClockRate / 1000);
    printf("   Memory Bus Width: %d bits\n", prop.memoryBusWidth);
    printf("   Peak Memory Bandwidth: %f GB/s\n\n",
           2.0 * prop.memoryClockRate * (prop.memoryBusWidth / 8) / 1.0e6);

    printf("10. DOUBLE PRECISION SUPPORT:\n");
    if (prop.major >= 1 && prop.minor >= 3) {
        printf("   Double Precision: Supported\n");
    } else {
        printf("   Double Precision: Not Supported\n");
    }
    printf("\n");

    printf("11. ADDITIONAL INFORMATION:\n");
    printf("   Clock Rate: %d MHz\n", prop.clockRate / 1000);
    printf("   Registers per Block: %d\n", prop.regsPerBlock);
    printf("   L2 Cache Size: %d KB\n", prop.l2CacheSize / 1024);
    printf("   Max Shared Memory per Multiprocessor: %zu KB\n", prop.sharedMemPerMultiprocessor / 1024);
    printf("   Concurrent Kernels: %s\n", prop.concurrentKernels ? "Yes" : "No");
    printf("   GPU Clock Rate: %d MHz\n\n", prop.clockRate / 1000);

    printf("12. MEMORY HIERARCHY:\n");
    printf("   L1 Cache Size per Multiprocessor: %zu KB\n", prop.l2CacheSize / 1024);
    printf("   Memory Bus Width: %d bits\n", prop.memoryBusWidth);
    printf("   Peak Bandwidth (GB/s): %.2f\n\n",
           2.0 * prop.memoryClockRate * (prop.memoryBusWidth / 8) / 1.0e6);

    printf("========================================\n");

    return 0;
}
EOF
```
